```python
import nltk     # First, we import the NLTK library to work with text data
from nltk.tokenize import wordpunct_tokenize   # This import helps us split a sentence into separate words
from nltk import pos_tag  # This function is used to assign grammar tags to words
```

```python
nltk.download('averaged_perceptron_tagger')  # We download the POS tagger model so NLTK knows how to tag words

nltk.download('averaged_perceptron_tagger_eng')    # This downloads the English language POS tagging data
```

```python
text = "Natural language processing is very interesting"   # Here we give the sentence that we want to analyze
```

```python
tokens = wordpunct_tokenize(text)  # Now we break the sentence into individual words
print(tokens)
```

```python
pos_tags = pos_tag(tokens)  # Next, we find the part of speech for each word
print(pos_tags)   # Finally, we print the words along with their POS tags
```
